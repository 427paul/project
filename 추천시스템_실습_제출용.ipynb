```python
import pandas as pd
import numpy as np
```

#Item-based CF(collaborative filtering)

```python
df = pd.merge(rating, movie[["id","title"]],on = "id")
df.head()
```

### 1.  df  item-based cf     df2  .  ( 0) <font color=blue>(10)</font>

```python
# -   
df2 = df.pivot_table(index='id', columns='userId', values='rating')

#  0 
df2 = df2.fillna(0)

# 
df2.head()
```

### 2. df2     cos_matrix_2     df2_cosine .<font color=blue>(10)</font><br>(,   df2 )

```python
from sklearn.metrics.pairwise import cosine_similarity

#     
cos_matrix_2 = cosine_similarity(df2)

#   
df2_cosine = pd.DataFrame(cos_matrix_2, index=df2.index, columns=df2.index)

# 
df2_cosine.head()
```

```python
rating.head()
```

```python
# 8:2 train test  

from sklearn.model_selection import train_test_split

train_rate, test_rate = train_test_split(rating, test_size=0.2, random_state=100)
```

### 3.item-based CF      
$$ \hat{r}_{5,i}  =  \frac{\sum_{j\in I_5}{sim(i,j)}\cdot r_{5,j}}{\sum_{j\in I_5}{sim(i,j)}} $$

* $\hat{r}_{5,i}$  5  i  
* $r_{5,j}$  5  j  
* $sim(i,j)$  i j 
* $I_5$  5     

#### 3-(1) train_rate  5  user5_train .<font color=blue>(5)</font>

```python
#train   5  
user5_train= train_rate[train_rate['userId'] == 5][['id', 'rating']]
user5_train.head()
```

#### 3-(2)     . <font color=blue>(20)</font>

```python
sim_sum = df2_cosine.loc[user5_train['id'].values, :].sum().values    # 
sim_sum = sim_sum + 1                                                # 0     +1 
sim_sum
```

#### 3-(3)    sim_rating .<font color=blue>(20)</font>

```python
sim_rating = np.matmul(df2_cosine.loc[user5_train['id'].values, :].T.values, user5_train['rating'].values)
sim_rating
```

#### 3-(4)  ,       pred_rating    .<font color=blue>(15)</font> <br>( df2_cosine )

```python
pred_rating = pd.DataFrame(np.divide(sim_rating, sim_sum), index=df2_cosine.index)
pred_rating.columns=["pred"]
pred_rating
```

### 4. test_rate 5  user5_test .<font color=blue>(5)</font>

```python
user5_test= test_rate[test_rate['userId'] == 5][['id', 'rating']]
user5_test.head()
```

### 5. user5_test pred_rating      .<font color=blue>(5)</font>

```python
item_cf = pd.merge(user5_test, pred_rating.loc[user5_test['id']], left_on = 'id',right_index=True)
item_cf.head()
```

### 6.      rmse, mae  .<font color=blue>(5)</font>

```python
import math
def evaluation(pred, rating):
    rmse=np.round(math.sqrt(np.mean((pred-rating)**2)),3)
    mae=np.mean(np.abs(pred-rating)).round(3)

    print('RMSE:', rmse)
    print('MAE:', mae)

evaluation(item_cf['rating'],item_cf['pred'])
```

```python
def similar_movie(title, cos_matrix):
    try:

        idx = indices[title]                #    ( ) 

        cos_sim = list(enumerate(cos_matrix[idx]))         #    
        cos_sim.sort(key=lambda x :x[1], reverse=True)    #    

        cos_sim = cos_sim[0:11]                           # 11  (     )
        sim_movie_idx=[x[0] for x in cos_sim]            #cos_sim   ,   
        sim_movie_idx.remove(idx)                        #  
        sim_movie_idx = sim_movie_idx[0:10]              # 10 


        title_list=movie['title'].iloc[sim_movie_idx]  #    
        title_list=pd.DataFrame(title_list)           # 

        id_list=movie['id'].iloc[sim_movie_idx]     #  id 
        id_list=pd.DataFrame(id_list)  #  

        temp1 = pd.concat([title_list,id_list], axis=1)   #title_list, id_list  
        temp2=pd.merge(temp1, pred_rating["pred"],left_on = 'id',right_index=True)    #temp1 pred_rating 

        result=temp2.drop(["id"], axis=1)
        result.rename(columns={"title":"  TOP10", "pred":""},inplace=True)

        print(result)


    except:
        idx=[]
        if len(indices[title])>1:

            for i in range(len(indices[title])):
                idx.append(indices[title][i])                          #   


                cos_sim = list(enumerate(cos_matrix[idx[i]]))         #    
                cos_sim.sort(key=lambda x :x[1], reverse=True)    #    


                cos_sim = cos_sim[0:11]                         # 11  (     )
                sim_movie_idx=[x[0] for x in cos_sim]            #cos_sim   ,   


                if idx[i] in sim_movie_idx:
                    sim_movie_idx.remove(idx[i])                    #  

                sim_movie_idx = sim_movie_idx[0:10]              # 10 



                title_list = movie['title'].iloc[sim_movie_idx]  #    
                title_list=pd.DataFrame(title_list)



                id_list=movie['id'].iloc[sim_movie_idx]     #  id 
                id_list=pd.DataFrame(id_list)


                temp1 = pd.concat([title_list,id_list], axis=1)

                temp2=pd.merge(temp1, pred_rating["pred"],left_on = 'id',right_index=True)    #temp1 pred_rating 



                result=temp2.drop(["id"], axis=1)
                result.rename(columns={"title":f" id:{idx[i]}   TOP10", "pred":""},inplace=True)

                print(result)
```
